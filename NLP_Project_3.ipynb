```python
!pip install googletrans==4.0.0.rc1

import googletrans
from googletrans import Translator

language_abbreviation_codes = {
    "French": "fr",
    "English": "en",
    "Italian": "it",
    "German": "de",
    "Spanish": "es",
    "Portuguese": "pt",
    "Dutch": "nl",
    "Russian": "ru",
    "Chinese": "zh",
    "Japanese": "ja",
    "Turkish": "tr",
    "Korean": "ko",
    "Arabic": "ar",
    "Hindi": "hi",
    "Swedish": "sv",
    "Danish": "da",
    "Norwegian": "no",
    "Finnish": "fi",
    "Polish": "pl",
    "Greek": "el",
    "Czech": "cs",
    "Ukrainian": "uk",
    "Bulgarian": "bg",
    "Catalan": "ca",
    "Hungarian": "hu",
    "Romanian": "ro",
    "Croatian": "hr"
}

def main():
    translator = Translator()

    print("Welcome to the NLP-Language Translator")

    source_text = input("Enter the text you want to translate: ")

    source_lang = choose_language("source")

    target_lang = choose_language("target")

    translated_text_parts = []
    max_text_length = 8000
    for i in range(0, len(source_text), max_text_length):
        part = source_text[i:i + max_text_length]
        translation_part = translate_text(translator, part, source_lang, target_lang)
        translated_text_parts.append(translation_part)

    full_translated_text = "\n".join(translated_text_parts)

    print("\nTranslation:")
    print(full_translated_text)

def choose_language(type):
    show_codes = input(f"Do you want to see the language codes for {type} language? (yes/no): ")
    if show_codes.lower() == "yes":
        print("\nAvailable languages:")
        for lang in language_abbreviation_codes:
            print(f"{lang}: {language_abbreviation_codes[lang]}")
    lang_input = input(f"Enter the {type} language: ")
    return language_abbreviation_codes.get(lang_input, lang_input)

def translate_text(translator, text, source_lang, target_lang):
    try:
        translation = translator.translate(text, src=source_lang, dest=target_lang)
        return translation.text
    except Exception as e:
        print("An error occurred:", str(e))
        return "Translation failed."

if __name__ == "__main__":
    main()
```
